```lua

Plot = require 'itorch.Plot'
require 'optim'

torch.setdefaulttensortype('torch.DoubleTensor')

function ftheta(x)
    theta0 = 2.5
    theta1 = -1.7
    theta2 = 3.1
    theta3 = -3.2
    theta4 = 1.9
    local fval = theta0 + theta1 *x*x + theta2 * torch.exp(theta3 * torch.abs(x) - theta4*x*x)
    return fval
end

trainx = torch.Tensor(1000):zero()
trainy = torch.Tensor(1000):zero()
xrange = torch.linspace(-2.5, 2.5, 1000)
for i=1, xrange:size(1) do
    trainx[i] = xrange[i]
    fexact = ftheta(xrange[i])
    noise = 0.5 - 2 * torch.rand(1)[1]
    trainy[i] = fexact + noise    
end

Plot():circle(trainx, trainy):title('Function graph'):draw()

-- COMPLETELY CLUELESS
-- ix = torch.randperm(1000)
-- ix = ix[{{1, 100}}]:chunk(100)
-- trainx = trainx[ix]
-- trainy = trainx:reshape(1, 1000)
-- trainy = trainy[{{}, ix}]

-- Plot():circle(trainx:resize(100), trainy:resize(100), 'green'):title('Train set'):draw()

-- 
ix = torch.randperm(1000)
ix = ix[{{1, 100}}]
newtrainx = torch.Tensor(100):zero()
newtrainy = torch.Tensor(100):zero()
for i = 1, 100 do
    newtrainx[i] = trainx[ix[i]]
    newtrainy[i] = trainy[ix[i]]
end
data = {
    inputs = newtrainx,
    targets = newtrainy
}

Plot():circle(data.inputs, data.targets, 'green'):title('Train set'):draw()
```

```lua

state = {
   learningRate = 1e-1,
   maxIter = 100,
}

d = 5
n = data.inputs:size(1)

theta = torch.randn(5)
grad = torch.randn(5)

function feval(x)
    -- local xval = x[1]
    local f = function(x)
        theta0 = x[1]
        theta1 = x[2]
        theta2 = x[3]
        theta3 = x[4]
        theta4 = x[5]

        local input = data.inputs
        local input2 = torch.Tensor(input:size()):zero()
        input2:addcmul(1, input, input)
        local absinput = torch.abs(input)

        local term1 = torch.mul(input2, theta1)
        local term2 = torch.mul(absinput, theta3)
        local term3 = torch.mul(input2, theta4)
        local term4 = torch.mul(torch.exp(term2 - term3), theta2)

        local fval = term1 + term4
        fval:add(theta0)

        -- local fval = term1
        -- fval:add(theta0)
        -- fval:mul(1/n)

        local sqloss = torch.norm(data.targets-fval)
        local loss = sqloss * sqloss
        loss = (1/n) * loss

        -- print('~~~~~~~~~~~~~~~~~~~')
        -- print(fval)
        -- print('~~~~~~~~~~~~~~~~~~~')

        return loss, fval
    end
 
    local g = function(x, z)
        theta0 = x[1]
        theta1 = x[2]
        theta2 = x[3]
        theta3 = x[4]
        theta4 = x[5]

        local input = data.inputs
        local input2 = torch.Tensor(input:size()):zero()
        input2:addcmul(1, input, input)
        local absinput = torch.abs(input)

        local gradtheta = torch.randn(d, n):zero()

        gradtheta[{{1}, {}}] = torch.ones(1, n)
        gradtheta[{{2}, {}}] = input2
        gradtheta[{{3}, {}}] = torch.exp(torch.mul(absinput, theta3) - torch.mul(input2, theta4))

        local term4_1 = torch.mul(absinput, theta3)
        local term4_2 = torch.mul(input2, theta4)
        local term4_3 = torch.exp(term4_1 - term4_2)
        local term4_4 = torch.Tensor(n):zero():addcmul(term4_3, absinput)
        gradtheta[{{4}, {}}]:mul(term4_4, theta2)

        local term5_1 = torch.Tensor(n):zero():addcmul(term4_3, input2)
        gradtheta[{{5}, {}}]:mul(term5_1, -theta2)

        -- gradtheta[{{5}, {}}]:addcmul(1, torch.exp(torch.mul(absinput, theta3) - torch.mul(input2, theta4)), input2)

        -- local loss, z = f(x)
        local gradloss  = -torch.mul(data.targets-z, 2)
        local gradloss  = gradloss:reshape(1, input:size(1)):expand(d, input:size(1))

        -- Be careful of the initializations
        local gradlosstheta = torch.Tensor(gradloss:size()):zero():addcmul(1, gradloss, gradtheta)

        local grad = torch.mean(gradlosstheta, 2)

        -- print('==================')
        -- print(z)
        -- print(gradlosstheta)
        -- print(grad)
        -- print('==================')

        return grad
    end
 
    local loss, z = f(x)
    local grad = g(x, z)

    return loss, grad
end
```

```lua

-- stop when the gradient is close to 0, or after many iterations
fall_ = torch.Tensor(50010):zero()
iter = 0
while true do
    -- optim has multiple functions, such as adagrad, sgd, lbfgs, and others
    -- see documentation for more details
    theta, f_ = optim.lbfgs(feval, theta, state)
    fall_[iter+1] = f_[1]
    -- print('f optimal: ' .. f_[1])
    -- gradient norm is SOMETIMES a good measure of how close we are to the optimum, but often not.
    -- the issue is that we'd stop at points like x=0 for x^3
    if grad:norm() < 0.005 or iter > 50000 then 
        break 
    end
    iter = iter + 1
end

print('f optimal: ' .. f_[1])
print('Optimal parameters:')
print(theta)
```

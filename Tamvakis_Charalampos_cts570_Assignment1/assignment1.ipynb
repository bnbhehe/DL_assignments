## Loss Function: 
The most common loss function for seeing how our approximated function is fitting the data is the Mean Squared Error Function : $ \mathcal{L}(x;\theta) = \frac 1 2 \frac 1 N \sum_\tau^T(f(x;\theta) - t_\tau)^2 $ <br>
For convenience with the derivatives we use the $\frac 1 2$ in front of the function.

## Function : 
$$ f(x; \boldsymbol\theta)=\theta_0 + \theta_1 x^2+ \theta_2 \exp(\theta_3 |x| - \theta_4 x^2).$$
## Partial derivatives for $\boldsymbol{\theta}$ of the function :
$${\nabla_{\theta_0} f(x; \boldsymbol \theta)}= {\partial f \over\partial \theta_0} = 1  \\ $$
$${\nabla_{\theta_1} f(x; \boldsymbol\theta)}= {\partial f \over\partial \theta_1} = x^2  \\ $$
$${\nabla_{\theta_2} f(x;\boldsymbol \theta)}= {\partial f \over\partial \theta_2} = \exp(\theta_3 |x| - \theta_4 x^2)  \\ $$
$${\nabla_{\theta_3} f(x;\boldsymbol \theta)}= {\partial f \over\partial \theta_3} = |x| \theta_2\exp(\theta_3 |x| - \theta_4 x^2)\\$$
$${\nabla_{\theta_4} f(x; \boldsymbol\theta)}= {\partial f \over\partial \theta_4} = -x^2\theta_2\cdot\exp(\theta_3 |x| - \theta_4 x^2)\\ 
$$

## So the gradient of the Loss Function would be 
$$\nabla_{\theta_0} \mathcal{L}(x;\boldsymbol\theta) = {\partial \mathcal{L} \over \partial \theta_0} = \frac 1 N \sum_\tau^T(f(x;\boldsymbol\theta) - t_\tau) \cdot {\partial f \over\partial \theta_0} = \frac 1 N \sum_\tau^T(f(x;\boldsymbol\theta) - t_\tau) $$
$$\nabla_{\theta_1} \mathcal{L}(x;\boldsymbol\theta) = {\partial \mathcal{L} \over \partial \theta_1} = \frac 1 N \sum_\tau^T(f(x;\boldsymbol\theta) - t_\tau) \cdot x^2
$$
$$\nabla_{\theta_2} \mathcal{L}(x;\boldsymbol\theta) = {\partial \mathcal{L} \over \partial \theta_2} = \frac 1 N \sum_\tau^T(f(x;\boldsymbol\theta) - t_\tau) \cdot \exp(\theta_3 |x| - \theta_4 x^2)   $$
$$\nabla_{\theta_3} \mathcal{L}(x;\boldsymbol\theta) = {\partial \mathcal{L} \over \partial \theta_3} = \frac 1 N \sum_\tau^T(f(x;\boldsymbol\theta) - t_\tau) \cdot |x| \theta_2\exp(\theta_3 |x| - \theta_4 x^2) $$
$$\nabla_{\theta_4} \mathcal{L}(x;\boldsymbol\theta) = {\partial \mathcal{L} \over \partial \theta_4} = \frac 1 N \sum_\tau^T(f(x;\boldsymbol\theta) - t_\tau) \cdot( -x^2\theta_2\cdot\exp(\theta_3 |x| - \theta_4 x^2))$$
